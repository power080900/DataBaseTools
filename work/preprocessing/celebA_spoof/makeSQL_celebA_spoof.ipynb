## query 작성

```python
query1 = (f'''CREATE TABLE IF NOT EXISTS celebA_spoof_info (
        id INT(11) NOT NULL AUTO_INCREMENT primary key,
        img_dir VARCHAR(100) NOT NULL,
        img_format VARCHAR(10) NOT NULL,
        img_width INT(11) NOT NULL,
        img_height INT(11) NOT NULL,
        color_space VARCHAR(10) NOT NULL,
        label_dir VARCHAR(100) NOT NULL,
        label_format VARCHAR(10) NOT NULL,
        label VARCHAR(10) NOT NULL,
        label_id INT(10) NOT NULL,
        box_xtl INT(10) NOT NULL,
        box_ytl INT(10) NOT NULL,
        box_w INT(10) NOT NULL,
        box_h INT(10) NOT NULL,
        box_score float(50) NOT NULL)''')
```

```python
query2 = (f'''INSERT INTO celebA_spoof_info
        (img_dir, img_format, img_width, img_height, color_space,
        label_dir, label_format, label, label_id, box_xtl, 
        box_ytl, box_w, box_h, box_score) VALUES 
          (%s, %s, %s, %s, %s,  %s, %s, %s, %s, %s,  %s, %s, %s, %s)''')
```

## qurey 실행

```python
try:
    doUT.create_table(query1)
except Exception as e:
    print(f"Error during table creation: {e}")
```

```python
label_path = []
label_name= []

# label 경로
label_path = 'Z:\\celebA_spoof\\data\\RGB\\labeldata'
image_path = 'Z:\\celebA_spoof\\data\\RGB\\rawdata'

value_list1 = []
error_list = []
count = 0

label_ids = os.listdir(label_path)
for num, label_id in enumerate(label_ids):
    total_cnt = len(label_ids)
    color_space = 'RGB'
    label_path_1 = os.path.join(label_path,label_id)
    labels = os.listdir(label_path_1)
    for label in labels:
        try:
            label_path_2 = os.path.join(label_path_1, label)
            label_files = os.listdir(label_path_2)
            for cnt, label_file in enumerate(label_files):


                label_name, label_format = os.path.splitext(os.path.basename(label_file))
                label_format = label_format.replace('.','')
                label_dir = os.path.join(label_path_2, label_file)
                if label_format == 'png':
                    os.remove(label_dir)
                    # print(label_dir)
                    continue
                with open(label_dir,'r', encoding='utf-8') as f:
                    text = f.read()

                img_dir = label_dir.replace('label','raw').replace('_BB.txt','.jpg')
                if not os.path.isfile(img_dir):
                    img_dir = label_dir.replace('label','raw').replace('_BB.txt','.png')


                img_file = os.path.basename(img_dir)
                img_name, img_format = img_file.split('.')

                if os.path.getsize(img_dir) != 0:
                    img = cv2.imread(img_dir)            
                else:
                    error_list.append(img_dir)
                    continue
                img_w = img.shape[0]
                img_h = img.shape[1]
                
                img_dir = img_dir.replace('Z:','DataBase')
                label_dir = label_dir.replace('Z:','DataBase')

                box_xtl, box_ytl, box_w, box_h, box_score = text.split(' ')
                box_xtl = int(float(box_xtl))
                box_ytl = int(float(box_ytl))
                box_w = int(float(box_w))
                box_h = int(float(box_h))

                box_score = box_score.replace('\n','')

                values = (img_dir, img_format, img_w, img_h, color_space,
                          label_dir, label_format, label, label_id, box_xtl, 
                          box_ytl, box_w, box_h, box_score)

                value_list1.append(values)
        except Exception as e:
            error_list.append(label_dir)
            print(label_dir, e)
            

        # query 실행
        if len(value_list1) >= 500:
            # print(value_list1[0])
            doUT.insert_dataset_values(query2, value_list1)
            value_list1 = []
            print(num, total_cnt)

doUT.insert_dataset_values(query2, value_list1)
# print(value_list2)
value_list1 = []
print('done')
```
