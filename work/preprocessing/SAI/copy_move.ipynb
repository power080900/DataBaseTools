```python
import shutil
import os
```

```python
ids = ['id_34']
```

```python
for id_ in ids:
    file_dir = f'Z:\\SAI\\org_data\\{id_}'
    save_dir = 'Z:\\SAI\\data\\IR'
    for file_case in os.listdir(file_dir):
        if "(2)" in file_case:
            # print(1)
            root_dir = f"{file_dir}\\{file_case}"
            print(root_dir)
            root_files = os.listdir(root_dir)
            target_case = file_case.replace("(2)","")
            tt = file_dir.replace("(2)","")
            target_dir = f"{tt}\\{target_case}"
            target_files = os.listdir(target_dir)
            target_num = int(len(target_files)/7)

            rep_path = f"{tt}\\{target_case}"
            print(rep_path)


            for root_file in root_files:
                rep_num = root_file.split('.')[0]
                rep_file = root_file.replace(f'{rep_num}.',f'{target_num}.',1)
                print(rep_file)
                rep = f'{rep_path}\\{rep_file}'
                root = f'{root_dir}\\{root_file}'

                shutil.move(root,rep)

                if '.segments' in root_file:
                    target_num += 1
```

```python
for id_ in ids:
    file_dir = f'Z:\\SAI\\org_data\\{id_}'
    save_dir = 'Z:\\SAI\\data\\IR'
    for case in os.listdir(file_dir):
        files = os.path.join(file_dir,case)
        for file in os.listdir(files):
            file_org = os.path.join(file_dir,case,file)
            img_rep = os.path.join(save_dir,'rawdata',f'{id_}',f'{case}_{file}')
            label_rep = os.path.join(save_dir,'labeldata',f'{id_}',f'{case}_{file}')
            seg_rep = os.path.join(save_dir,'segdata',f'{id_}',f'{case}_{file}')

            if not os.path.exists(os.path.dirname(img_rep)):
                os.makedirs(os.path.dirname(img_rep))
            if not os.path.exists(os.path.dirname(label_rep)):
                os.makedirs(os.path.dirname(label_rep))
            if not os.path.exists(os.path.dirname(seg_rep)):
                os.makedirs(os.path.dirname(seg_rep))

                
            try:
                if '.rgb' in file:
                    shutil.copy(file_org,img_rep)

                elif '.info' in file:
                    shutil.copy(file_org,label_rep)

                elif '.segment' in file:
                    shutil.copy(file_org,seg_rep)
            except:
                print(file_org)
```
