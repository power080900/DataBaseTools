```python
import cv2
import numpy as np
import os
import sys
sys.path.append('../../src')
import DISData as DD
import ast
```

```python
doUT = DD.SQL()
```

```python
df = doUT.search_SQL('celebA_box')
```

```python
df_box = df[['img_dir','box_xtl','box_ytl','box_w','box_h']]
```

```python
df_box
```

```python
for index, row in df.iterrows():
    img = cv2.imread(row['img_dir'])
    xtl, ytl, w, h = row['box_xtl'], row['box_ytl'], row['box_w'], row['box_h']
    xbr, ybr = xtl + w, ytl + h
    cv2.rectangle(img, (xtl, ytl), (xbr, ybr), (0, 255, 0), 2)  # 바운딩 박스 그리기
    cv2.imshow('Image with bounding box', img)
    cv2.waitKey(0)  # 키보드 입력 대기 (아무 키나 누를 때까지)
    cv2.destroyAllWindows()  # 창 닫기
```

```python
df2 = doUT.search_SQL('celebA_landmark')
df_lan = df2[['img_dir','landmark']]
df_lan
```

```python
for index, row in df_lan.iterrows():
    img = cv2.imread(row['img_dir'])
    landmarks = ast.literal_eval(row['landmark'])  # 문자열로 된 튜플을 파이썬 객체로 변환
    for landmark in landmarks:
        x, y = landmark
        cv2.circle(img, (x, y), 3, (0, 0, 255), -1)  # 빨간색 원 그리기
    cv2.imshow('Image with landmarks', img)
    cv2.waitKey(0)  # 키보드 입력 대기 (아무 키나 누를 때까지)
    cv2.destroyAllWindows()  # 창 닫기
```
