```python
data_name = 'zer01ne'
```

```python
query = f'''select img_name, img_dir, label, box_xtl, box_ytl, box_xbr, box_ybr from DeepInSight.{data_name}_box_info;'''
```

```python
db = doUT.db
conn = doUT.conn
conn.execute(query)
column_names = [desc[0] for desc in conn.description]
result = conn.fetchall()
db.commit()
```

```python
df = pd.DataFrame(result, columns=column_names)
```

```python
import cv2
import os

def draw_boxes_on_image(df, save_dir):
    # 이전 이미지 파일 이름 초기화
    prev_img_name = None
    
    for index, row in df.iterrows():
        print(prev_img_name)
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = 0.5
        color = (255, 255, 255)  # White color
        thickness = 1
        img_name = row['img_name']
        img_dir = row['img_dir'].replace('DataBase','Z:')
        img_path = os.path.join(img_dir)
        
        
        if img_name != prev_img_name:
            # 이미지 파일 경로 생성
            
            
            # 이미지 저장
            if not os.path.exists(save_dir):
                os.makedirs(save_dir)
            save_path = os.path.join(save_dir, f'{img_name}.jpg')
            # print(save_path)
            image = cv2.imread(img_path)
            cv2.imwrite(save_path, image)
            prev_img_name = img_name
            
        else:
            label = row['label']
            box_xtl = row['box_xtl']
            box_ytl = row['box_ytl']
            box_xbr = row['box_xbr']
            box_ybr = row['box_ybr']
            print((box_xtl, box_ytl), (box_xbr, box_ybr))
            org = (box_xtl, box_ytl - 10)

            # 이미지 불러오기
            

            # 상자 그리기
            image = cv2.rectangle(image, (box_xtl, box_ytl), (box_xbr, box_ybr), (0, 255, 0), 2)
            image = cv2.putText(image, label, org, font, font_scale, color, thickness, cv2.LINE_AA)
            cv2.imshow('Bounding Box with Label', image)
            cv2.waitKey(0)
            cv2.destroyAllWindows()

            prev_img_name = img_name
```

```python
save_dir = 'Z:\zer01ne\data\IR\chechdata'  # 이미지를 저장할 디렉토리

draw_boxes_on_image(df, save_dir)
```
