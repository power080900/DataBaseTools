## SQL 연결

```python
doUT = DD.SQL()
```

```python
# query 실행
query1 = (f'''CREATE TABLE IF NOT EXISTS ETH_Xgaze_info (
        id INT(11) NOT NULL AUTO_INCREMENT primary key,
        img_dir VARCHAR(100) NOT NULL,
        img_format VARCHAR(10) NOT NULL,
        img_width INT(11) NOT NULL,
        img_height INT(11) NOT NULL,
        color_space VARCHAR(10) NOT NULL,
        label_dir VARCHAR(100) NOT NULL,
        label_format VARCHAR(10) NOT NULL,
        cam_index VARCHAR(10) NOT NULL,
        face_mat_norm VARCHAR(1000) NOT NULL,
        Gaze_2D VARCHAR(1000) NOT NULL,
        Head_2D VARCHAR(1000) NOT NULL)''')
```

```python
query2 = (f'''INSERT INTO ETH_Xgaze_info 
        (img_dir, img_format, img_width, img_height, color_space,
          label_dir, label_format, cam_index, face_mat_norm, 
          Gaze_2D, Head_2D) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s )''')
```

```python
try:
    doUT.create_table(query1)
except Exception as e:
    print(f"Error during table creation: {e}")
```

```python
# label 경로
label_path = os.path.join("Z:\\", "ETH_Xgaze", "data", "RGB","labeldata","train")
image_path = os.path.join("Z:\\", "ETH_Xgaze", "data", "RGB","rawdata","train")

value_list1 = []
error_list = []

label = os.path.join(label_path, 'train.label')
try:
    with open(label, 'r') as file:
        texts = file.readlines()
        for line in texts[1:]:
            face, gaze, head, origin, cam_index, frame_index, normmat = line.split(' ')
            img_dir = image_path + "/" + str(face).replace("\\", "/")
            img_format = 'jpg'
            image = cv2.imread(img_dir)
            img_width = image.shape[1]
            img_height = image.shape[0]
            color_space = 'RGB'
            label_dir = label
            label_format = 'label'
            Gaze_2D = gaze
            Head_2D = head
            cam_index = cam_index
            face_mat_norm = normmat
            img_dir = img_dir.replace('Z:', 'DataBase')
            label_dir = label_dir.replace('Z:', 'DataBase')
            value = (img_dir, img_format, img_width, img_height, color_space, label_dir, label_format, cam_index, face_mat_norm, Gaze_2D, Head_2D)
            # print(value)
            value_list1.append(value)

            # query 실행
            if len(value_list1) >= 500:
            # print(value_list1[0])
                # print(f'landmark {cnt} / {total_cnt}')
                doUT.insert_dataset_values(query2, value_list1)
                value_list1 = []

except Exception as ex:
    error_list.append((label, str(ex)))
    print(f"Error loading data from file: {label}")
    print(f"Error details: {ex}")
    
doUT.insert_dataset_values(query2, value_list1)
# print(value_list2)
value_list1 = []
print('done')    
```
