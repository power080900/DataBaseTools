```python
# import libraries
import numpy as np
from pathlib import Path
from random import randint
import matplotlib.pyplot as plt
from typing import Tuple, List
from json import loads
```

```python
def project_point_to_screen(projection: np.array, point: np.array, resx: int = 640, resy: int = 360) -> Tuple[int, int]:
    M = np.ones((4, 4))
    M[0, :] = point[0]
    M[1, :] = point[1]
    M[2, :] = point[2]
    M *= projection
    vtemp = np.sum(M, axis=0)

    if vtemp[3] > 0:
        invW = 1 / vtemp[3]
        return int((0.5 + vtemp[0] * 0.5 * invW) * resx), int((0.5 - vtemp[1] * 0.5 * invW) * resy)
    return 0, 0


def get_projected_points(data_file: Path) -> List[Tuple[int, int]]:
    out: List[Tuple[int, int]] = []
        
    # load annotation file 
    json_data = loads(data_file.read_text())
    
    # projected keypoints coordinates array agnostic to keypoint
    view_projection_mat: np.array = np.array(json_data["Camera"]["ViewProjection"])
    
    for key_data in json_data["Keypoints"].values():
        out.append(project_point_to_screen(view_projection_mat, key_data))
        
    return out
```

```python
import pandas as pd
import json
import os

base_path = "nsmount/Sysnthetic Human Eyes/data/RGB/labeldata"

#  DataFrame    
total_label_data = pd.DataFrame()

#  :
# for label_name in tqdm(os.listdir(base_path), desc="processing Image files"):
for label_name in os.listdir(base_path):
    label_path = os.path.join(base_path, label_name)
    if label_name.endswith(".json"):  # : file_name  label_name 
        # JSON   
        with open(label_path, 'r') as file:
            data = json.load(file)

        # 'Overall'  
        if 'Overall' in data:
            gaze_data = data['Overall'].get("GazeImage", [])

            # 'look_vec'   DataFrame 
            look_vec_x = gaze_data[0] if gaze_data else None
            look_vec_y = gaze_data[1] if len(gaze_data) > 1 else None

            # DataFrame 
            label_data = pd.DataFrame({'look_vec_x': [look_vec_x], 'look_vec_y': [look_vec_y]})

            #  total_label_data DataFrame 
            total_label_data = pd.concat([total_label_data, label_data], ignore_index=True)
        else:
            print(f"Key 'Overall' not found in {label_path}")

#  DataFrame 
total_label_data
```

```python
# Convert radians to degrees
total_label_data['look_vec_x_deg'] = np.degrees(total_label_data['look_vec_x'])
total_label_data['look_vec_y_deg'] = np.degrees(total_label_data['look_vec_y'])
```

## joint plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'total_label_data' DataFrame is already created from your previous code

# Plotting the jointplot
sns.jointplot(data=total_label_data, x='look_vec_x_deg', y='look_vec_y_deg', kind='kde', cmap='RdYlGn')
plt.xlabel('2DGaze_X',fontsize=10)
plt.ylabel('2DGaze_Y',fontsize=10)
plt.figure(figsize=(16, 12))
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'total_label_data' DataFrame is already created from your previous code

cmap = sns.color_palette("YlGnBu", as_cmap=True)

# Create a 2D histogram with annotations
hist_data, x_edges, y_edges, im = plt.hist2d(total_label_data['look_vec_x_de'], total_label_data['look_vec_y_deg'], bins=10, cmap=cmap)

# Annotate each bin with its count
for i in range(len(x_edges) - 1):
    for j in range(len(y_edges) - 1):
        count = int(hist_data[i, j])
        if count > 0:
            x_center = (x_edges[i] + x_edges[i + 1]) / 2
            y_center = (y_edges[j] + y_edges[j + 1]) / 2
            if count > 30000 : 
                plt.text(x_center, y_center, str(count), color='white',
                    ha='center', va='center', fontsize=8)
            else :
                plt.text(x_center, y_center, str(count), color='black',
                    ha='center', va='center', fontsize=8)
# Add a color bar
cbar = plt.colorbar(im)

plt.xlabel('2DGaze_X',fontsize=10)
plt.ylabel('2DGaze_Y',fontsize=10)
plt.figure(figsize=(16, 12))
```
