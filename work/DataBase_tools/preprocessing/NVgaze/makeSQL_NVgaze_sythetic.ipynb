```python
query1 = (f'''CREATE TABLE NVgave_synthetic_dataset_info (
        id INT(11) NOT NULL AUTO_INCREMENT primary key,
        
        img_dir VARCHAR(500) NOT NULL,
        label_dir VARCHAR(500) NOT NULL,
        label_format VARCHAR(10) NOT NULL,
        archive VARCHAR(100) NOT NULL,
        image VARCHAR(200) NOT NULL,
        
        regionmask_withskin VARCHAR(200) NOT NULL,
        regionmask_withoutskin VARCHAR(200) NOT NULL,
        eye VARCHAR(10) NOT NULL,
        eye_x VARCHAR(300) NOT NULL,
        
        eye_y VARCHAR(300) NOT NULL,
        eye_z VARCHAR(300) NOT NULL,
        blink  VARCHAR(300) NOT NULL,
        gaze_x_degree VARCHAR(300) NOT NULL,
        gaze_y_degree VARCHAR(300) NOT NULL,
        
        iris_texture_f VARCHAR(300) NOT NULL,
        pupil_center_x_4mm_dilation VARCHAR(300) NOT NULL,
        pupil_center_x_6mm_dilation VARCHAR(300) NOT NULL,
        pupil_center_y_4mm_dilation VARCHAR(300) NOT NULL,
        pupil_center_y_6mm_dilation VARCHAR(300) NOT NULL,
        
        pupil_size VARCHAR(300) NOT NULL,
        slippage_x VARCHAR(300) NOT NULL,
        slippage_y VARCHAR(300) NOT NULL,
        slippage_z VARCHAR(300) NOT NULL,
        iris_texture_i VARCHAR(300) NOT NULL);''')
try:
    doUT.create_table(query1)
except Exception as e:
    print(f"Error during table creation: {e}")
```

```python
query2 = (f'''INSERT INTO NVgave_synthetic_dataset_info 
        (img_dir, label_dir, label_format, archive, image, 
        regionmask_withskin, regionmask_withoutskin, eye, eye_x, eye_y,
        eye_z, blink, gaze_x_degree, gaze_y_degree, iris_texture_f, 
        pupil_center_x_4mm_dilation, pupil_center_x_6mm_dilation, pupil_center_y_4mm_dilation, pupil_center_y_6mm_dilation, pupil_size,
        slippage_x, slippage_y, slippage_z, iris_texture_i) VALUES 
        ( %s, %s, %s, %s, %s,  %s, %s, %s, %s, %s,  %s, %s, %s, %s, %s,  %s, %s, %s, %s, %s,  %s, %s, %s, %s )''')
```

```python
value_list1 = []
error_list = []
csv_dir = 'Z:\\NVgaze_synthetic_dataset\\data\\labeldata'
label_ids = os.listdir(csv_dir)

for num, label_id in enumerate(label_ids):
    csv_path = f"{csv_dir}\\{label_id}\\footage_description.csv"
    data = pd.read_csv(csv_path)
    for i in range(len(data['archive'])):
        img_dir = csv_path.replace('labeldata','rawdata').replace('footage_description.csv',data['image_L_0'][i])
        label_dir = csv_path
        label_format = os.path.basename(csv_path).split('.')[-1]
        label_dir = label_dir.replace('Z:','DataBase')
        img_dir = img_dir.replace('Z:','DataBase')
        archive = data['archive'][i]
        image = data['image_L_0'][i]
        regionmask_withskin = data['regionmask_withskin_L_0'][i]
        regionmask_withoutskin = data['regionmask_withoutskin_L_0'][i]
        eye = data['string_eye'][i]
        eye_x = data['string_original_eye_x'][i]
        eye_y = data['string_original_eye_y'][i]
        eye_z = data['string_original_eye_z'][i]
        blink = data['float_blink'][i]
        gaze_x_degree = data['float_gaze_x_degree'][i]
        gaze_y_degree = data['float_gaze_y_degree'][i]
        iris_texture_f = data['float_iris_texture_orientation'][i]
        pupil_center_x_4mm_dilation = data['float_pupil_center_displacement_x_at_4mm_dilation'][i]
        pupil_center_x_6mm_dilation = data['float_pupil_center_displacement_x_at_6mm_dilation'][i]
        pupil_center_y_4mm_dilation = data['float_pupil_center_displacement_y_at_4mm_dilation'][i]
        pupil_center_y_6mm_dilation = data['float_pupil_center_displacement_y_at_6mm_dilation'][i]
        pupil_size = data['float_pupil_size'][i]
        slippage_x = data['float_slippage_x'][i]
        slippage_y = data['float_slippage_y'][i]
        slippage_z = data['float_slippage_z'][i]
        iris_texture_i = data['int_iris_texture_index'][i]
        
        value = (img_dir, label_dir, label_format, archive, image, regionmask_withskin,
        regionmask_withoutskin, eye, eye_x, eye_y, eye_z,  blink, gaze_x_degree, gaze_y_degree, iris_texture_f, pupil_center_x_4mm_dilation,
        pupil_center_x_6mm_dilation, pupil_center_y_4mm_dilation, pupil_center_y_6mm_dilation, pupil_size, slippage_x,
        slippage_y, slippage_z, iris_texture_i)
        # print(value)
        value_list1.append(value)

        # query 실행

        if len(value_list1) >= 500:
            doUT.insert_dataset_values(query2, value_list1)
            value_list1 = []
            print(f'{num} / {len(label_ids)}')
            
if len(value_list1) != 0:
    doUT.insert_dataset_values(query2, value_list1)
    # print(value_list2)
    value_list1 = []
    print('done')      
else:
    value_list1 = []
    print('done')
```
