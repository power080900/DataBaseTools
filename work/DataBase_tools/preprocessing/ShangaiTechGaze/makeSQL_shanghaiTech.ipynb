## query 작성

```python
query1 = (f'''CREATE TABLE shanghaiTechGaze_info (
        id INT(11) NOT NULL AUTO_INCREMENT primary key,
        img_dir VARCHAR(100) NOT NULL,
        img_format VARCHAR(10) NOT NULL,
        img_width INT(11) NOT NULL,
        img_height INT(11) NOT NULL,
        color_space VARCHAR(10) NOT NULL,
        label_dir VARCHAR(100) NOT NULL,
        label_format VARCHAR(10) NOT NULL,
        which_eye VARCHAR(100) NOT NULL,
        cam_loc VARCHAR(100) NOT NULL,
        xy_gt VARCHAR(100) NOT NULL,
        tablet_w VARCHAR(100) NOT NULL,
        tablet_h VARCHAR(100) NOT NULL,
        between_cam VARCHAR(100) NOT NULL,
        landmark VARCHAR(1000) NOT NULL
        )''')
```

```python
query2 = (f'''INSERT INTO shanghaiTechGaze_info 
        (img_dir, img_format, img_width, img_height, color_space, 
        label_dir, label_format, which_eye, cam_loc, xy_gt, 
        tablet_w, tablet_h, between_cam, landmark) VALUES 
        (%s, %s, %s, %s, %s,  %s, %s, %s, %s, %s,  %s, %s, %s, %s)''')
```

## query 실행

```python
try:
    doUT.create_table(query1)
except Exception as e:
    print(f"Error during table creation: {e}")
```

```python
co_label_path = 'Z:\\shanghaiTechGaze\\data\\RGB\\labeldata\\coordinate'
lan_label_path = 'Z:\\shanghaiTechGaze\\data\\RGB\\labeldata\\face_landmarks'
image_path = 'Z:\\shanghaiTechGaze\\data\\RGB\\rawdata'


value_list1 = []
error_list = []

cam_locs = os.listdir(lan_label_path)
target_ids = os.listdir(co_label_path)

for cam_loc in cam_locs:
    for target_id in target_ids:
        label_files = os.listdir(os.path.join(co_label_path,target_id))
        for label_file in label_files:
            co_label = io.loadmat(os.path.join(co_label_path,target_id,label_file))
            lan_label = io.loadmat(os.path.join(lan_label_path,cam_loc,target_id,label_file))
            
            left_img_file = label_file.replace('.mat',"_left.jpg")
            right_img_file = label_file.replace('.mat',"_right.jpg")
            img_dir1 = os.path.join(image_path, cam_loc, target_id, left_img_file)
            img_dir2 = os.path.join(image_path, cam_loc, target_id, right_img_file)
            # print(img_dir)
            img_format = 'jpg'
            image1 = cv2.imread(img_dir1)
            image2 = cv2.imread(img_dir2)
            img_width = image1.shape[1]
            img_height = image1.shape[0]
            color_space = 'RGB'
            label_dir_co = os.path.join(co_label_path,target_id,label_file)
            label_dir_lan = os.path.join(lan_label_path,cam_loc,target_id,label_file)
            label_format = 'mat'
            img_dir1 = img_dir1.replace('Z:','DataBase').replace('\\','/')
            img_dir2 = img_dir2.replace('Z:','DataBase').replace('\\','/')
            label_dir = label_dir_co.replace('Z:','DataBase').replace('\\','/')
            which_eye1 = img_dir1.split('_')[-1].split('.')[0]
            which_eye2 = img_dir2.split('_')[-1].split('.')[0]
            xy = co_label['xy_gt'].tolist()
            xy_gt = ','.join(map(str,(float(xy[0][0]), float(xy[1][0]))))
            
            tablet_w = 59.77
            tablet_h = 33.62
            between_cam = 18.22
            lk = lan_label['eyelocation'].tolist()
            landmark = []
            for n,i in enumerate(lk):
                if n % 2 == 1:
                    y = i[0]
                    landmark.append((x,y))
                else:
                    x = i[0]
            landmark = ','.join(map(str,[(x,y) for x,y in landmark]))
            # print(landmark)
            value1 = (img_dir1, img_format, img_width, img_height, color_space, label_dir_co, label_format, which_eye1, cam_loc, xy_gt, tablet_w, tablet_h, between_cam, landmark)
            # print(value)
            value_list1.append(value1)
            value2 = (img_dir2, img_format, img_width, img_height, color_space, label_dir_co, label_format, which_eye2, cam_loc, xy_gt, tablet_w, tablet_h, between_cam, landmark)
            value_list1.append(value2)

            # query 실행

            if len(value_list1) >= 500:
                # print(value_list1[0])
                doUT.insert_dataset_values(query2, value_list1)
                value_list1 = []
                
doUT.insert_dataset_values(query2, value_list1)
# print(value_list2)
value_list1 = []
print('done')
```
